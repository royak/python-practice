```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
pi = np.pi
x = np.linspace(-4*pi, 4*pi, 100)
```

```python
y = np.sin(x)
```

```python
plt.plot(x,y)
plt.show()
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
plt.plot(x,sigmoid(x),c='red',lw=10,ls=':')
plt.grid()
plt.show()
```
